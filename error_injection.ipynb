```python
import pandas as pd
import numpy as np

df = pd.read_csv("Telco-Customer-Churn.csv")
df_errors = df.copy()

# 1. Missing values in a required column
df_errors.loc[5:20, "tenure"] = np.nan

# 2. Unknown categorical values
df_errors.loc[0:10, "InternetService"] = "Satellite" 

# 3. Wrong numeric values
df_errors.loc[15:25, "MonthlyCharges"] = -50 

# 4. Wrong categorical values
df_errors.loc[30:35, "gender"] = "child" 

# 5. String in numeric column
df_errors.loc[40:45, "tenure"] = "ten"

# 6. Extreme outliers
df_errors.loc[50:55, "MonthlyCharges"] = 9999

# 7. Drop a required column
df_with_missing_column = df_errors.drop(columns=["TotalCharges"], errors="ignore")

# Save one combined dataset with all errors
df_with_missing_column.to_csv("telco_with_errors.csv", index=False)

print("Error dataset generated at data/errors/telco_with_errors.csv")
```
